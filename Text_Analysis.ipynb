```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Load dataset
df = pd.read_csv('/content/aircraft_incident_200.csv')
df.head()
```

✅ This transformation turns free-form text into a structured format that can be:

Used in machine learning models (like clustering)

Analyzed for frequent causes

Visualized using word clouds, bar charts, or heatmaps



```python
import nltk
nltk.download('stopwords')

import re
from nltk.corpus import stopwords
from nltk.tokenize import TreebankWordTokenizer

stop_words = set(stopwords.words('english'))
tokenizer = TreebankWordTokenizer()

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = tokenizer.tokenize(text)
    filtered_tokens = [w for w in tokens if w not in stop_words]
    return ' '.join(filtered_tokens)

df['cleaned_text'] = df['cause_description'].apply(preprocess_text)
df[['cause_description', 'cleaned_text']].head()
```

```python
text = ' '.join(df['cleaned_text'])

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Top Keywords in Aircraft Incident Causes', fontsize=20)
plt.show()
```

Create one word cloud for each cluster to visualize the most common words in each group of aircraft incident causes.



```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Number of clusters (assumes 'cluster' column exists)
num_clusters = df['cluster'].nunique()

# Set up subplots
fig, axes = plt.subplots(1, num_clusters, figsize=(20, 6), squeeze=False)
fig.suptitle('✈️ Word Clouds by Cluster - Aircraft Incident Causes', fontsize=18)

# Loop through clusters and generate a word cloud for each
for cluster_id in range(num_clusters):
    cluster_text = ' '.join(df[df['cluster'] == cluster_id]['cleaned_text'])
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(cluster_text)

    ax = axes[0][cluster_id]
    ax.imshow(wordcloud, interpolation='bilinear')
    ax.axis('off')
    ax.set_title(f'Cluster {cluster_id}', fontsize=14)

plt.tight_layout(rect=[0, 0, 1, 0.93])  # Adjust for suptitle
plt.show()
```

```python
# Vectorization
vectorizer = TfidfVectorizer(max_df=0.8, min_df=2)
X = vectorizer.fit_transform(df['cleaned_text'])

# KMeans Clustering
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X)

df['cluster'] = kmeans.labels_
df[['cause_description', 'cluster']].head()
```

```python
sns.countplot(data=df, x='cluster')
plt.title("Incident Causes Cluster Distribution")
plt.xlabel("Cluster")
plt.ylabel("Count")
plt.show()
```

```python
from collections import Counter

for i in range(5):
    print(f"\nCluster {i} Top Words:")
    cluster_texts = df[df['cluster'] == i]['cleaned_text']
    words = ' '.join(cluster_texts).split()
    most_common = Counter(words).most_common(10)
    print(most_common)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Use CountVectorizer to get word counts
vectorizer = CountVectorizer(max_features=20)  # Limit to top 20 keywords
X_counts = vectorizer.fit_transform(df['cleaned_text'])

# Create a DataFrame for the word counts
word_count_df = pd.DataFrame(X_counts.toarray(), columns=vectorizer.get_feature_names_out())
word_count_df['cluster'] = df['cluster']

# Group by cluster and sum the word counts
cluster_word_freq = word_count_df.groupby('cluster').sum()

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(cluster_word_freq.T, annot=True, cmap="YlGnBu", fmt='g')
plt.title("🔥 Heatmap of Keyword Frequency by Cluster", fontsize=16)
plt.xlabel("Cluster")
plt.ylabel("Keyword")
plt.tight_layout()
plt.show()
```
